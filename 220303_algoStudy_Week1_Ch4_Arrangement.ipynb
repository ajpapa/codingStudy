# Selection Sort
 - For each iteration, compare elements to find the biggest element among them.
 - Place the biggest element at the back of compared elements.
 - For n elements, (n-1) iteration is needed.
 - In k-iteration(1<=k<=n-1), (n-k) comparison occur.
 - O(n**2)

```python
#Selection Sort
def selectionSort(A:list) -> list :
    n=len(A)
    for i in range(n-1): # n-1 iteration
        idMax=0
        print(f"i iteration for i = {i}")
        
        for j in range(1,n-i) : # n-k comparision
            print(f"    j iteration for j = {j}")
            
            if A[j]>A[idMax] :
                idMax=j #update                
            print(f"        idMax is {idMax} and it's value is {A[idMax]}")
            
        A[idMax], A[n-i-1] = A[n-i-1], A[idMax] # place the biggest element at the back
        print(f"after {i}-iteration A is {A}")
        
    return A
```

```python
A=[31,8,48,73,11,3,20,29,65,15]
print(f"length of A is {len(A)}")
selectionSort(A)
```

# Bubble Sort
 - For each iteration(n-1), compare every adjacent elements and switch each other if they were not sorted.
 - For n elements, (n-1) iteration is needed.
 - In k-iteration(1<=k<=n-1), (n-k) comparison occur.
 - O(n**2)

```python
def bubbleSort(A:list) -> list :
    n=len(A)
    
    for i in range(n-1):
        print(f"i iteration for i = {i}")
        
        for j in range(n-i-1):
            print(f"    j iteration for j = {j}")
            if A[j]>A[j+1] :
                A[j], A[j+1] = A[j+1], A[j]
            print(f"    after {j}-iteration A is {A}")
        print(f"after {i}-iteration A is {A}")
                
    return (A)
        
```

```python
A=[31,8,48,73,11,3,20,29,65,15]
print(f"length of A is {len(A)}")
bubbleSort(A)
```

# Insertion Sort
 - Build orders by putting each unordered element to the appropriate place among ordered elements
 - O(n**2)

```python
def insertionSort(A:list)->list :
    n=len(A)
    for i in range(1,n) : #두번째 원소부터 배치 시작
        print(f"i iteration for i = {i}")

        for j in range(i,0,-1):
            print(f"    j iteration for j = {j}")
            
            if A[j-1] > A[j] :
                A[j-1], A[j] = A[j], A[j-1]
            print(f"    after j={j}-iteration A is {A}")

        print(f"after i={i}-iteration A is {A}")

    return A
```

```python
A=[31,8,48,73,11,3,20,29,65,15]
print(f"length of A is {len(A)}")
insertionSort(A)
```

# Merge Sort
 - recursive method
 - idea : divide a list into two lists, sort each lists and then merge that two lists
 - O(nlogn)

```python
def mergeSort(A:list, p:int, q:int)->list :
    # the end of recursion
    if q-p==1 :# when length is 1 -> return itself
        return A[p:q]
    
    else :
        # divide list
        r=p+(q-p)//2
        print(f"p, r, q = {p}, {r}, {q}")

        # for each list call mergeSort
        left=mergeSort(A, p, r)
        print(f"LEFT for p={p},~ r={r}, q={q} is {left}")
        right=mergeSort(A, r, q)
        print(f"RIGHT for p={p}, r={r},~ q={q} is {right}")

        # merge that two lists
        id_left=0
        id_right=0
        i=0
        while ( (id_left<r-p) and (id_right<q-r) ) :
            if left[id_left]<=right[id_right] :
                A[p+i]=left[id_left]
                id_left+=1
            else :
                A[p+i]=right[id_right]
                id_right+=1
            i+=1
        if id_left==r-p :
            A[p+i:q]=right[id_right:q-r]
        elif id_right==q-r :
            A[p+i:q]=left[id_left:r-p]

        print(f"after function call A[p:q] for p={p}, r={r},q={q} is {A[p:q]}")

        return A[p:q]
```

```python
A=[31,8,48,73,11,3,20,29,65,15]
mergeSort(A,0,len(A))
```

# Quick Sort
 - idea 
          pick one element among list(ex : the lattest one) 
          and then with respect to that element(let's call it pivot),
          elements smaller than pivot are arranged on the left of pivot,
          and elements larger than pivot are arranged on the right of pivot
          and do this for left of pivot and right of pivot again and again(recursive)
 - There exits base element(i.e. pivot) for each iteration
 - best performance in average O(nlogn)
 - But, for the worst case O(n**2) since T(n)=T(n-1)+theta(n) for that case

```python
#quick selection

def quickSort(A : list, p : int, r : int )-> None :
    if p<r :
        q = partition(A, p, r)
        print(f"After PARTITION(A={A},p={p},r={r}) is called q ={q} ")
        quickSort(A, p, q-1)
        print(f"After QuickSort(A={A},p={p},q-1={q-1}) is called A[p:q] is ={A[p:q]} ")
        quickSort(A, q+1, r)
        print(f"After QuickSort(A={A},q+1={q+1},r={r}) is called A[q+1:r] is ={A[q+1:r+1]} ")
    else :
        return
    
def partition(A : list, p : int, r : int) -> int :
    pivot=A[r]
    print(f"In partition for p~r = {p}~{r}, pivot = A[{r}] = {pivot}")
    i=p-1
    for j in range(p,r) :
        if A[j]<=pivot :
            i+=1 # i는 비교원소가 pivot보다 작으면 계속 증가, pivot보다 크면 큰 원소 바로 직전 원소에 멈춰있음
            A[i], A[j]=A[j], A[i] #뒤쪽에서 pivot보다 작은 element 찾으면 큰 원소(i)와 작은 원소(j)를 바꿔버림
    q=i+1 #pivot이 들어갈 위치
    A[q],A[r]=A[r], A[q]
    print(f"After partition is done for p~r = {p}~{r}, A[{p}:{r+1}] is {A[p:r+1]}")
    return q
```

# Heap Sort
 - heap structure in array : binary tree / ordering btwn ancestor and decestors / left to right


```python
#heap sort - 최소 heap 으로-


# Ftn making heap structure /  n : last index
def buildHeap(A : list, n : int)->None :
    for i in range( (n-1)//2,-1,-1) :
        heapify(A, i, n)
    return


# Ftn making heap structure under A[k] / k:corresponding node / n : last index
def heapify(A:list, k : int, n : int) -> None : 

    #A[k]와 하위 노드(A[2*k+1], A[2*k+2]) 간의 관계 정리. (단, 하위 노드가 n 이내임을 확인하기)
    if 2*k+1 > n : #하위 노드가 없으면 return
        return
    elif 2*k+ 1 == n : #하위 노드가 1개이면 비교 대상으로 선정
        smaller, s_index = A[2*k+1], 2*k+1
    else : # 하위 노드가 2개이면 비교 대상은 작은 것으로 선정
        if A[2*k+1] <= A[2*k+2] :
            smaller, s_index = A[2*k+1], 2*k+1
        else : 
            smaller, s_index = A[2*k+2], 2*k+2

            
    if smaller < A[k] : # A[k] 노드가 하위 노드보다 값이 크면 
        A[k], A[s_index] = A[s_index], A[k] # 서로 바꾸고,
        heapify(A, s_index, n) #하위 노드도 heapify 해준다
    return

def heapSort(A:list, n: int)-> list :
    buildHeap(A, n)
    print("after building Heap:", A)
    for i in range(n,0,-1):
        A[0], A[i] =A[i], A[0]
        heapify(A,0, i-1)
        print("A :",A, "when iteration is done with i :", i)
    
```

```python
templist=[7,9,4,8,6,3]
heapSort(templist,len(templist)-1)
print(templist)
```

# leet code - 905. Sort Array By Parity
# link : https://leetcode.com/problems/sort-array-by-parity/

```python
def sortArrayByParity(nums : list)-> list:
    """
    :type nums: List[int]
    :rtype: List[int]
    """
    n=len(nums)
    if n>1:
        i=-1
        for j in range(n):
            if nums[j]%2==0 :
                i+=1
                nums[i], nums[j] =nums[j], nums[i]
    return nums
```

# leet code - 506. Relative Ranks
# link : https://leetcode.com/problems/relative-ranks/

```python
# leet code - 506. Relative Ranks
# link : https://leetcode.com/problems/relative-ranks/

class Solution(object):
    def findRelativeRanks(self, score):
        """
        :type score: List[int]
        :rtype: List[str]
        """
        
        #ordering scores       

        temp=[] #for matching btwn 'score' and 'rank' // 
        for num in score:
            temp.append(num)
        #주의 ★★★★ list 변수를 temp로 만들 때는 그냥 같다고 하면,,temp를 변경하는 행위가 본래 list를 변경하는 결과를 초래

        self.quickSort(temp, 0, len(temp)-1) #reordering in descending order
#         print("temp after sorting is", temp)
#         print("score after sorting is", score)


        #matching btwn 'score' and 'rank'
        matDic=dict() #matching dictionary
        for i, num in enumerate(temp) :
            matDic[num]=i+1 #key is 'score' and value is 'rank'
#         print("matDic is", matDic)        


        #building rank
        rank=[]
        for num in score:
            rank.append(matDic[num])
#             print("num is", num, "and rank is", rank)
#         print("rank is", rank)        
        
        #building result
        result=[]
        for num in rank:
            if num==1:
                result.append("Gold Medal")
            elif num==2:
                result.append("Silver Medal")
            elif num==3:
                result.append("Bronze Medal")
            else:
                result.append(str(num))
#         print("result is", result)        
        return result

    
#     @classmethod
    def quickSort(self, A : list, p : int, r : int ) -> None :
        if p<r :
            q = self.partition(A, p, r)
            self.quickSort(A, p, q-1)
            self.quickSort(A, q+1, r)
        else :
            return
        
#     @classmethod
    def partition(self, A : list, p : int, r : int) -> int :
        base=A[r]
        i=p-1
        for j in range(p,r) :
            if A[j]>=base :
                i+=1
                A[i], A[j]=A[j], A[i]
        q=i+1
        A[q],A[r]=A[r], A[q]
        return q
```

```python
sol=Solution()
score = [10,3,8,9,4,1]
sol.findRelativeRanks(score)
```

# leet code - 2164. Sort Even and Odd Indices Independently
# link : https://leetcode.com/problems/sort-even-and-odd-indices-independently/
#odd indices non-increasing
#even indices non-decreasing

```python
# leet code - 2164. Sort Even and Odd Indices Independently
# link : https://leetcode.com/problems/sort-even-and-odd-indices-independently/
#odd indices non-increasing
#even indices non-decreasing

class Solution(object):
    def sortEvenOdd(self, nums):
        """
        :type nums: List[int]
        :rtype: List[int]
        """
        #odd indices ordering change by insertion sort
        
        for i in range(3,len(nums),2):
            loc=i
            while ((loc>1) and (nums[loc]>=nums[loc-2])) :
                nums[loc], nums[loc-2]=nums[loc-2], nums[loc]
                loc-=2
                
        #even indices ordering change by insertion sort
        
        for i in range(2,len(nums),2):
            loc=i
            while ((loc>0) and (nums[loc]<=nums[loc-2])) :
                nums[loc], nums[loc-2]=nums[loc-2], nums[loc]
                loc-=2
        return nums
        
```

```python
sol=Solution()
nums = [4,1,2,3]
sol.sortEvenOdd(nums)
```

# 767. Reorganize String `Medium`
* [https://leetcode.com/problems/reorganize-string](https://leetcode.com/problems/reorganize-string)

```python
class Solution(object):
    def reorganizeString(self, s):
        """
        :type s: str
        :rtype: str
        """
        from collections import Counter
        counter=Counter(s)
        picked=None 
        ans='_' # _ is sentinel
        
        while counter :
            
            #가장 많은 2개의 문자 중, 직전에 입력한 것과 다른 문자를 선택
            for ele, num in counter.most_common(2):
                if ele!=ans[-1] : 
                    picked=ele
                    break
            print(f"counter.most_common(2) is ★{counter.most_common(2)}")
            print(f"picked is ★{picked}")
            # 선택한 문자를 추가
            if picked!=None:
                ans+=picked
                print(f"updated ans is ★{ans}")
                # 사용후 사용내역을 반영
                counter[picked]-=1 #사용 후 개수 제거
                if counter[picked]<=0 :
                    counter.pop(picked)
                print(f"after adding, counter is ★{counter}")
                #picked 업데이트
                picked=None
            else : #pick이 안 되면 끝냄
                break

        ans=ans[1:]
        if len(ans)==len(s) :
            return ans
        else :
            return ""
                
```

# 767. Reorganize String
# https://leetcode.com/problems/reorganize-string/
# idea is by size of the frequent

```python
"""
Squeezing number betweens the characters with maximum counts returns
the string without any adjacent characters
"""


class Solution:
    def reorganizeString(self, s: str) -> str:
        s = list(s)
        counting_dict = dict()
        loop = 0

        # 1 dictionary whose key and value is character and count each.
        for elem in s:
            try:
                counting_dict[elem] += 1
            except KeyError:
                counting_dict[elem] = 1

        # 2 list is reformatted into the list which is sorted by counts of characters.
        while counting_dict:
            key = max(counting_dict, key=counting_dict.get)
            for j in range(0, counting_dict[key]):
                s[loop + j] = key
            loop += counting_dict[key]
            del counting_dict[key]

        max_count = s.count(s[0])

        """#3 if the maximum value of count is greater than that of others, 
        there's no means to get desirable results."""

        if loop - max_count < max_count - 1:
            return ""
        else:
            for i in range(len(s) - max_count):
                if not i % max_count:
                    max_idx = [i for i, value in enumerate(s) if value == s[0]]
                    print("if not works for i :", i, 'and max_idx is : ', max_idx)
                if s[max_count + i] != s[max_idx[0]]:
                    s.insert(
                        max_idx[i % max_count] + i % max_count + 1, s.pop(max_count + i)
                    )
                    print("if works for i :", i, 'and max_count+i is : ', max_count+i, 'with max_idx[0] is',max_idx[0], 'and after that, s is', s)

        return "".join(s)
```
