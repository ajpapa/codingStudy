# 평균적으로 O(n)인 selection 알고리즘
 - Quick Sort시 사용한 partition을 활용하여 search할 범위를 줄이는 방법
 - 최악의 경우에는 O(n**2)이 됨

```python
# A[p : r+1] 에서 i 번째로 작은 원소를 찾는 함수
def selectQuick(A:list, p:int, r:int, i:int) -> int: 

    if p==r :
        if i==1 :
            return A[p]
        else :
            return 99999
    else :
        q=partition(A,p,r)
        #A[r] 값을 기준으로 
        #왼쪽에는 해당 값보다 작은 element, 
        #오른족에는 큰 element가 배열되며, 
        #A[r]값의 새로운 위치(q)가 return


        if (q-p)==(i-1) :
            return A[q]
        #만약 q 앞에 i-1개가 있으면, 
        #A[q]가 i번째로 작은 값이므로 함수 종료

        else :
            if (q-p) < (i-1) : #q 우측에서 찾아야 함
                return selectQuick(A,q+1,r,i-(q-p+1))
            else : # q-p > i-1 #q 좌측에서 찾아야 함
                return selectQuick(A,p,q-1,i)

            
    
def partition(A : list, p : int, r : int) -> int :
    pivot=A[r]
#     print(f"In partition for p~r = {p}~{r}, pivot = A[{r}] = {pivot}")
    i=p-1
    for j in range(p,r) :
        if A[j]<=pivot :
            i+=1 # i는 비교원소가 pivot보다 작으면 계속 증가, pivot보다 크면 큰 원소 바로 직전 원소에 멈춰있음
            A[i], A[j]=A[j], A[i] #뒤쪽에서 pivot보다 작은 element 찾으면 큰 원소(i)와 작은 원소(j)를 바꿔버림
    q=i+1 #pivot이 들어갈 위치
    A[q],A[r]=A[r], A[q]
#     print(f"After partition is done for p~r = {p}~{r}, A[{p}:{r+1}] is {A[p:r+1]}")
    return q
```

# O(n) selection algorithm even in Worst Case
 - key idea : preprocessing in picking pivot element(divide elements with its size 5 and use median of medians)

 - other reference
  - https://blog.naver.com/babobigi/221828095048
  - https://rcoh.me/posts/linear-time-median-finding/

```python
def linearSelect(A:list, p:int, r:int, i:int) -> int:

    #원소개수가 5개 이하며 바로 찾기
    if p-r+1<=5 :
        return selectQuick(A,p,r,i)

    
    #5개씩 나누었을 때 그룹 수   
    if (p-r+1)%5 ==0 :
        groupNum=(p-r+1)/5
    else :
        groupNum=(p-r+1)//5+1

        
    #각 그룹의 median값 모을 변수 준비
    medians=[0]*groupNum

    
    #각 그룹의 median값 구하기
    for i in range(groupNum-1) :
        medians[i]=selectQuick(A, p+(5*i), p+(5*i)+4, 3)
    medians[-1]=selectQuick(A, p+5*(groupNum-1),r, ((p-r+1)%5) //2) #마지막 그룹만 별도로 연산

    #recursive 하게 median들의 median 구하기
    
    ## median들의 median은 몇번째 index인지 구하기
    if len(medians)%2==0:
        idMedian=medians//2
    else:
        idMedian=medians//2+1
    
    #★recursive하게 median of medians 구하기
    mediOfMedi=linearSelect(medians, 0, len(medians)-1,idMedian)

    
    # mediOfMedi를 pivot으로 하여 분할
   
    pivot=mediOfMedi
    k=p-1
    for j in range(p,r+1) :
        if A[j]<=pivot :
            k+=1 # k는 비교원소가 pivot보다 작거나 같으면 계속 증가, pivot보다 크면 큰 원소 바로 직전 원소에 멈춰있음
            A[k], A[j]=A[j], A[k] #뒤쪽에서 pivot보다 작거나 같은 element 찾으면 큰 원소(k)와 작은 원소(j)를 바꿔버림
    #정렬은 끝났고,
    #pivot이 위치한 index는 k가 됨
    
    #pivot 좌우 중 적합한 쪽 선택
    if k-p+1==i:
        return A[k]
    elif k-p+1<i:
        return linearSelect(A, k+1, r, i-(k-p+1))
    else : # k-p+1>i
        return linearSelect(A, p, k-1, i)
    
```

# 740. Delete and Earn(Medium)
 - https://leetcode.com/problems/delete-and-earn/

        You are given an integer array nums. 
        You want to maximize the number of points you get by performing the following operation any number of times:

        Pick any nums[i] and delete it to earn nums[i] points. 
        Afterwards, you must delete every element equal to nums[i] - 1 and every element equal to nums[i] + 1.
        Return the maximum number of points you can earn by applying the above operation some number of times.



        Example 1:
        Input: nums = [3,4,2]
        Output: 6
        Explanation: You can perform the following operations:
        - Delete 4 to earn 4 points. Consequently, 3 is also deleted. nums = [2].
        - Delete 2 to earn 2 points. nums = [].
        You earn a total of 6 points.

        Example 2:
        Input: nums = [2,2,3,3,3,4]
        Output: 9
        Explanation: You can perform the following operations:
        - Delete a 3 to earn 3 points. All 2's and 4's are also deleted. nums = [3,3].
        - Delete a 3 again to earn 3 points. nums = [3].
        - Delete a 3 once more to earn 3 points. nums = [].
        You earn a total of 9 points.


        Constraints:
        1 <= nums.length <= 2 * 10**4
        1 <= nums[i] <= 10**4

```python
class Solution(object):
    def deleteAndEarn(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        #point dictionary 생성
        maxVal=max(nums)
        point=dict(zip(range(1,maxVal+1),[0]*maxVal))
        for num in nums:
            point[num]+=num
        
        ans=[0]*(maxVal+1) #ans 담을 변수
        ans[1]=point[1]
        
        for i in range(2, maxVal+1) :
            ans[i]=max(ans[i-2]+point[i], ans[i-1])
        return ans[maxVal]
```

```python
sol=Solution()
templist = [2,2,3,3,3,4]
sol.deleteAndEarn(templist)
```

# ★Following is Sangsoo's solution(Top Down Dynamic Programming)

```python
class Solution(object):
    def deleteAndEarn(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        #point dictionary 생성
        point=dict(zip(range(1,105),[0]*104))
        for num in nums:
            point[num]+=num
        
        
        #list에서 가장 큰 값을 찾고
        maxVal=max(nums)
        return findMax(maxVal)
        
        #ftn for Top-down DP / point dictionary is given
        def findMax(maxVal:int)->int:
            if maxVal==0 :
                return 0
            elif maxVal==1 :
                return point(maxVal)
            else : #maxVal>=2
                return max( findMax(maxVal-2) + point[maxVal] , findMax(maxVal-1) )
```

# ★Following is Namju's solution(divide and conquer strategy)

```python
# divide-and-conquer solution, not dynamic programming

nums = [3,4,2] # 6
nums = [2,2,3,3,3,4] # 9
nums = [8,7,3,8,1,4,10,10,10,2] # 52
nums = [3,7,10,5,2,4,8,9,9,4,9,2,6,4,6,5,4,7,6,10] # 66
    # (2 * 2) + (4 * 4) + (6 * 3) + (8 * 1) + (10 * 2) = 66
# nums = [2,2,2,3,3,3,4,4,4,5,5,6]

nums.sort() # sort the data
key, weight, comp, c = [], [], [], 0 
# key: sorted unique list of keys(not essential); 
# weight: how much we can earn deleting a key entirely; 
# comp: starting indices of consecutive chunks
prev = -999
for n in nums:
    if n != prev: # when a new number appears for the first time
        key.append(n) # record the appearance of the new number
        weight.append(n) # create a placeholder for the new number with trivial unit weight
        if n > prev + 1: # if consecutiveness doesn't hold
            comp.append(c) # add a brand new consecutive chunk
        c += 1 # since a brand new chunk appears, c increases
    else: # when the current number is the same with the previous one
        weight[-1] += n # add weight to the previously created placeholder
    prev = n

# find the best earning for the range [start, end)
def find(weight, start, end): # start: inclusive; end: exclusive
    if end - start <= 0:
        return 0
    elif end - start == 1: # when the range length is 1
        return weight[start] # trivially return itself
    elif end - start == 2: # when the range length is 2
        return max(weight[start], weight[start + 1]) # return larger one
    else: # when the range length is larger than 3
        mid = (end + start) // 2
        mid_incl = weight[mid] + find(weight, start, mid - 1) + find(weight, mid + 2, end) # taking mid for earning then call recursively
        mid_excl = find(weight, start, mid) + find(weight, mid + 1, end) # not taking mid for earning and call recursively
        return max(mid_incl, mid_excl) # return larger one(taking mid or not)

sum_ = 0
for i in range(len(comp)): # for each consecutive component(chunk)
    # determine [start, end) range
    start = comp[i]
    try:
        end = comp[i + 1]
    except IndexError:
        end = len(weight)
    # find the best earning for the current consecutive component
    ret = find(weight, start, end)
    sum_ += ret
    print(f'find(weight, {start}, {end}) -> {ret} {sum_}')

print(sum_)
# return sum_
```

# 324. Wiggle Sort II (Medium)

 - https://leetcode.com/problems/wiggle-sort-ii/
 - Given an integer array nums, reorder it such that nums[0] < nums[1] > nums[2] < nums[3]....
 - You may assume the input array always has a valid answer.

 

 - Example 1:
    - Input: nums = [1,5,1,1,6,4]
    - Output: [1,6,1,5,1,4]
    - Explanation: [1,4,1,5,1,6] is also accepted.
 - Example 2:
    - Input: nums = [1,3,2,2,3,1]
    - Output: [2,3,1,3,1,2]
 

 - Constraints:
    - 1 <= nums.length <= 5 * 104
    - 0 <= nums[i] <= 5000
    - It is guaranteed that there will be an answer for the given input nums.
 

 - Follow Up: Can you do it in O(n) time and/or in-place with O(1) extra space?

```python
class Solution(object):
    def wiggleSort(self, nums):
        """
        :type nums: List[int]
        :rtype: None Do not return anything, modify nums in-place instead.
        """
        nums.sort(reverse=True)
        even=nums[:len(nums)//2]
        odd=nums[len(nums)//2:]
        nums[1::2]=even
        nums[::2]=odd
        return nums
    
sol=Solution()
templist = [10,3,8,9,9,4,1]
sol.wiggleSort(templist)
```

# 347. Top K Frequent Elements(Medium)
 - https://leetcode.com/problems/top-k-frequent-elements/
 - Given an integer array nums and an integer k, return the k most frequent elements. You may return the answer in any order.
 
 - Example 1:
    - Input: nums = [1,1,1,2,2,3], k = 2
    - Output: [1,2]
 - Example 2:
    - Input: nums = [1], k = 1
    - Output: [1]
 - Constraints:
    - 1 <= nums.length <= 105
    - k is in the range [1, the number of unique elements in the array].
    - It is guaranteed that the answer is unique.

```python
class Solution(object):
    def topKFrequent(self, nums, k):
        """
        :type nums: List[int]
        :type k: int
        :rtype: List[int]
        """
        #by using counter
        from collections import Counter
        ans=[ num for num,counts in Counter(nums).most_common(k)]
        return ans
        
sol=Solution()
templist = [10,8,3,3,9,3,3,8,9,9,4,1]
print(sol.topKFrequent(templist,3))
```

# 973. K Closest Points to Origin(Medium)

 - Given an array of points where points[i] = [xi, yi] represents a point on the X-Y plane and an integer k, return the k closest points to the origin (0, 0).
 - The distance between two points on the X-Y plane is the Euclidean distance (i.e., √(x1 - x2)2 + (y1 - y2)2).
 - You may return the answer in any order. The answer is guaranteed to be unique (except for the order that it is in).
 - Example 1:
            - Input: points = [[1,3],[-2,2]], k = 1
            - Output: [[-2,2]]
            - Explanation:
            - The distance between (1, 3) and the origin is sqrt(10).
            - The distance between (-2, 2) and the origin is sqrt(8).
            - Since sqrt(8) < sqrt(10), (-2, 2) is closer to the origin.
            - We only want the closest k = 1 points from the origin, so the answer is just [[-2,2]].
 - Example 2:
            - Input: points = [[3,3],[5,-1],[-2,4]], k = 2
            - Output: [[3,3],[-2,4]]
            - Explanation: The answer [[-2,4],[3,3]] would also be accepted.


 - Constraints:
        - 1 <= k <= points.length <= 104
        - -104 < xi, yi < 104

```python
class Solution(object):
    def kClosest(self, points, k):
        """
        :type points: List[List[int]]
        :type k: int
        :rtype: List[List[int]]
        """
        lengthList=[]
        for x,y in points:
            lengthList.append((x**2+y**2))
        print(f"lengthList is {lengthList}")
        temp=sorted(zip(lengthList, points))
        print(f"temp is {temp}")
        sol=[]
        for i in range(k) :
            sol.append(temp[i][1])
        return sol

sol=Solution()
templist = [[3,3],[5,-1],[-2,4]]
print(sol.kClosest(templist,2))    
```

# 1985. Find the Kth Largest Integer in the Array(Medium)
 - https://leetcode.com/problems/find-the-kth-largest-integer-in-the-array/
 
        You are given an array of strings nums and an integer k.
        Each string in nums represents an integer without leading zeros.
        Return the string that represents the kth largest integer in nums.

        Note: 
        Duplicate numbers should be counted distinctly. 
        For example, if nums is ["1","2","2"], 
        "2" is the first largest integer, "2" is the second-largest integer,
        and "1" is the third-largest integer.

        Example 1:

        Input: nums = ["3","6","7","10"], k = 4
        Output: "3"
        Explanation:
        The numbers in nums sorted in non-decreasing order are ["3","6","7","10"].
        The 4th largest integer in nums is "3".


        Example 2:

        Input: nums = ["2","21","12","1"], k = 3
        Output: "2"
        Explanation:
        The numbers in nums sorted in non-decreasing order are ["1","2","12","21"].
        The 3rd largest integer in nums is "2".


        Example 3:

        Input: nums = ["0","0"], k = 2
        Output: "0"
        Explanation:
        The numbers in nums sorted in non-decreasing order are ["0","0"].
        The 2nd largest integer in nums is "0".

        Constraints:
        1 <= k <= nums.length <= 104
        1 <= nums[i].length <= 100
        nums[i] consists of only digits.
        nums[i] will not have any leading zeros.

```python
class Solution(object):
    def kthLargestNumber(self, nums, k):
        """
        :type nums: List[str]
        :type k: int
        :rtype: str
        """
        temp=[]
        for i, num in enumerate(nums):
            temp.append([int(num), i])
        temp.sort()
        return nums[temp[-k][1]]
            
sol=Solution()
templist = ["2","21","12","1"]
print(sol.kthLargestNumber(templist,3)) 
```
